## Bee swarm plots

```python
# to avoid hist cons use bee swarm plots
import seaborn as sns
sns.set()

_ = sns.swarmplot(x = 'state', y = 'dem_share', data=df_swing)
_ = plt.xlabel('State')
_=plt.ylabel('% of vote for Obama')
plt.show()
```

```python
# Beeswarm has limitation (overlapping data points)
# hence, using beeswarm is not the best option here

# Alternative- ESDF(Empirical cumulative distribution function)
```

```python
import numpy as np
x = np.sort(df_swing['dem_share'])
y = np.arange(1, len(x)+1) / len(x)
_ = plt.plot(x, y, marker='.', linestyle='none')
_ = plt.xlabel('% of vote for Obama')
_ = plt.ylabel('ECDF')
# plt.margins(0.02) # keeps data off plot edges
plt.show()
```

```python
import numpy as np
x = np.sort(df_swing['dem_share'])
y = np.arange(1, len(x)+1) / len(x)
_ = plt.plot(x, y, marker='.', linestyle='none')
_ = plt.xlabel('% of vote for Obama')
_ = plt.ylabel('ECDF')
plt.margins(0.02) # keeps data off plot edges
plt.show()
```

### plotting multiple ECDF usinf iris dataset

```python
from sklearn.datasets import load_iris
data = load_iris()
type(data)
```

```python
iris_data = pd.DataFrame(np.c_[data['data'], data['target']],
            columns= data['feature_names'] + ['target'])

# pd.DataFrame(np.concatenate((data.data, np.array([data.target]).T), axis=1),
#             columns = data.feature_names + ['target'])
```

```python
iris_data.head()
```

```python
def ecdf(data):
    x = np.sort(data)
    y = np.arange(1, len(data)+1)/len(data)
    
    return x, y

# data.target_names
```

```python
x_set, y_set = ecdf(iris_data[iris_data['target']==0]['petal length (cm)'])
x_vers, y_vers = ecdf(iris_data[iris_data['target']==1]['petal length (cm)'])
x_virg, y_virg = ecdf(iris_data[iris_data['target']==2]['petal length (cm)'])
```

```python
plt.plot(x_set, y_set, marker='.', linestyle='none')
plt.plot(x_vers, y_vers, marker='.', linestyle='none')
plt.plot(x_virg, y_virg, marker='.', linestyle='none')
plt.xlabel('Petal length (cm)')
plt.ylabel('ECDF')
plt.margins(0.02)


# Annotate legends
plt.legend(('setosa','versicolor','virginica'),
          loc = 'lower right')
plt.show()
```
